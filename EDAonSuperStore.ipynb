### Dealing with the Missing Data in the dataset

```python
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
df.describe()
```

## Category

```python
df.groupby(["Category"]).mean()
```

```python
df.groupby(["Category"]).max()
```

```python
df.groupby(["Category"]).min()
```

```python
df.groupby(["Category"]).describe()
```

```python
df.groupby('Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.show()
```

## Sub-Category

```python
df.groupby('Sub-Category').mean()
```

```python
df.groupby('Sub-Category').max()
```

```python
df.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.show()
```

```python
# Correlation in Dataset
correlation=df.corr()
correlation
```

```python
# Heat Map Visualization
sns.heatmap(correlation,annot=True,cmap='hot')
```

```python
# Ploting based upon the REGION and the category based.
plt.figure(figsize=(25,10))
plt.subplot(2,1,1)
sns.countplot('Region',hue='Category',data=df)
plt.title('REGION-CATEGORY')
plt.ylabel('COUNT',fontsize=11)
plt.xlabel('REGION',fontsize=11)
plt.grid(alpha=0.3)
plt.show()
```

```python
# Plotting based upon the REGION and the Sub-category based.
plt.figure(figsize=(25,10))
plt.subplot(2,1,2)
sns.countplot('Region',hue='Sub-Category',data=df)
plt.title('REGIONS-SUB CATEGORY')
plt.ylabel('COUNT',fontsize=12)
plt.xlabel('REGION',fontsize=10)
plt.grid(alpha=0.3)
plt.show()
```

```python
sns.countplot(x=df['Ship Mode'])
```

```python
plt.figure(figsize=(20,15))
df['Sub-Category'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

```python
# Sales per Ship mode
df.groupby('Ship Mode')['Sales','Profit'].agg(['sum']).plot.bar()
plt.title('Year wise Total Sales & % of profit gained')
plt.xticks(rotation=0)
```

```python
# Checking fo the state wise number of delivery.
plt.figure(figsize=(15,7.5))
sns.countplot(x=df['State'])
plt.xticks(rotation=90)
plt.show()
```

```python
top_state=df.groupby('State').sum().sort_values('Profit',ascending=False)
top_state.head(15)
```

```python
# Pairplot
figsize=(16,10)
sns.pairplot(df,hue='Sub-Category')
```
