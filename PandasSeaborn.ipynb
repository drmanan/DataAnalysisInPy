```python
width=24
plt.figure(figsize=(width,6))
col = 'State'
temp=pd.DataFrame(df[col].value_counts().sort_values(ascending=False)[:35],columns=[col])
sns.barplot(data=temp,x=temp.index,y=col,palette='tab10')
plt.xlabel(col)
plt.xticks(rotation=60)
```

```python
# Count plot counts, Bar plot takes mean, count plot takes either X or Y, not both 
col = 'Sub-Category'
width=24
plt.figure(figsize=(width,6))
temp=pd.DataFrame(df[col].value_counts().sort_values(ascending=False)[:35],columns=[col])
sns.countplot(data=temp,x=temp.index,y=col,palette='tab10')
plt.xlabel(col)
plt.xticks(rotation=60)
```
